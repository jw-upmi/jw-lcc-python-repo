AW 03 - Write a while loop that asks the user to enter two numbers. The numbers should be added and the sum displayed. The loop should ask the user if he or she wishes to perform the operation again. If so, the loop should repeat, otherwise it should terminate.

```python
print('==================')
print('= ADDING NUMBERS =')
print('==================')
# define keep_going to start the loop
keep_going = 'y'
# while keep_going is yes than loop through
while keep_going == 'y':
    # prompt user for two numbers.  found input().split will allow input to captuer more than one input at a time as a string.
    num1, num2 = input('Input two numbers: ').split(' ')
    # print input numbers
    print(f'You entered {num1} and {num2}')
    # add num1 and num2
    sum = int(num1) + int(num2)
    # print sunn
    print(f'The sum of {num1} and {num2} is {sum}')
    # prompt user to keep_going
    keep_going = input('Do you wish to add two more numbers? (y/n) ')
```

AW 05 - Write a loop that calculates the total of the following series of numbers: [1/30+2/29+3/28+...+30/1]

```python
print('========================')
print('= 1-30 DIVIDED by 30-1 =')
print('========================')
print('\nThis is a loop that will add 1/30+2/29+3/28+...+30/1.\n')
# set number of iterations
iteration = 30
# predefine numerator to be 0
numerator = 1
# predefine denominator to be equal to number of iterations
denominator = iteration
# predefinte carryover to be 0
carryover = 0 
# wetup while
while numerator <= iteration:
    result = (numerator/denominator)
    carryover += result
    print(f'\t{numerator}/{denominator} = {result} which equals {carryover}')
    numerator += 1
    denominator -= 1
print(f'\n1/30+2/29+3/28+...+30/1 equals {round(carryover, 2)}.')
```

PE 01 - Bug Collector: A bug collector collects bugs every day for five days. Write a program that keeps a running total of the number of bugs collected during the five days. The loop should ask for the number of bugs collected for each day, and when the loop is finished, the program should display the total number of bugs collected.

```python
print('=====================')
print('==== BUG TRACKER ====')
print('=====================')
# predefine starting variables
ttlBugs = 0 # total bugs
addBugs = 0 # bugs added this day
day = 1
ttlDays = 5 
# check to ensure day is not more than 5 
while day < 6:
    # while day is < 6 and range starts at 1 and goes to ttlDay + 1
    for day in range(1, ttlDays + 1):
        # prompt user for number of bugs collected today
        addBugs = int(input(f'How many bugs did you collect on day {day}? '))
        # add addBugs to running ttlBugs
        ttlBugs += addBugs
        # print results
        print(f'\tYou collected {addBugs} on day # {day}.  You currently have {ttlBugs} bugs.')
        # increase day by 1
        day += 1
print(f'You collected a total of {ttlBugs} bugs in {ttlDays} days!  Excellent work!')
```

PE 14 - Write a program that uses nested loops to draw this pattern:
\n  *******
\n  ******
\n  *****
\n  ****
\n  ***
\n  **
\n  *

```python
print('======================')
print('= REMOVE AN ASTERISK =')
print('======================')
# set size of asterisk triangle
size = 7
# for loop for row in reverse count
for row in range(size, 0, -1):
    # print the number of astericks for each column
    for column in range(row):
        print('*', end='')
        row -= 1
```
