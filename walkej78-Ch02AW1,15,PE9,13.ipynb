###\#15. Write the turtle graphics statement to draw a square that is 100 pixels wide on each side and a circle that is centered inside the square.  The circle's radius should be 80 pixels.  The circle should be filled with the color red.  (the square should not be filled with color.)
####NOTE:  This will not display in Colab headless environment but will display on a machine with.

```python
import turtle           # import the turtle library
turtle.hideturtle()     # hide the turtle cursor
turtle.speed(2)         # slow the turtle cursor
turtle.penup()          # stop drawing
turtle.goto(50, 50)     # move turle cursor to 50,50 on the x,y axis
turtle.pendown()        # start drawing
turtle.goto(50, -50)    # move from 50,50 to 50,-50 -- 100 pixels
turtle.goto(-50, -50)   # move from 50,-50 to -50,-50 -- 100 pixels
turtle.goto(-50, 50)    # move from -50,-50 to -50,50 -- 100 pixels
turtle.goto(50, 50)     # move from 50,50 to 50,50 -- 100 pixels
turtle.penup()          # stop drawing
turtle.goto(0, -40)     # move from 50,50 to 0,-40 to draw a 40 pixel diameter circle -- 80 pixel radius
turtle.pendown()        # start darwing
turtle.fillcolor('red') # declare fill to red
turtle.begin_fill()     # start fill (red)
turtle.circle(40)       # created a 40 pixel diamter circle -- 80 pixel radius
turtle.end_fill()       # stop fill
turtle.done()           # keep drawing onscreen
```

###\#9. Celsius to Fahrenheit Temperature Converter
####Write a program that converts Celsius temperatures to Fahrenheit temperatures.  The formula is as follows:  F=((9/5)*C)+32.  

####The program should ask the user to enter a temperautre in Celsius, then display the temperature converted to Fahrenheit.

```python
print('=================================================')
print('= CELSIUS TO FAHRENHEIGHT TEMPERATURE CONVERTER =')
print('=================================================\n')
temp_C = float(input(print('Please input your temperature in Celsius: ')))  # Prompts user for temp in C as a float variable temp_C.
temp_F = ((9/5)*temp_C)+32                                                  # Converts temp_C variable to F as temp_F.
print('\nYou entered your temperature in Celsius as ', temp_C, '.', sep='') # Prints users C input.
print('Your temperature in Fahrenheit is', temp_F)                          # Print C input as F.
```

###\#13. Planting Grapevines
####A vinyard owner is planting several new rows of grapevines, and needs to know how many grapvines to plant in each row.  She has determined that after measuring the length of a future row, she can use the following formula to calculate the number of vines that will fit in the row, along with the trellis end-post assemblies that will need to be contucted at each end of row:  V=(R-(2*E))/S.  

####The terms in the formula are:
####V is the number of grapevines that will fit in the row.
####R is the length of the row, in feet.
####E is the amount of space, in feet, used by an end-post assmebly.
####S is the space between vinces, in feet.

####Write a program that makes the calculation for the vinyard owner.  The program should ask the user to input the following:
####1. The length of the row, in feet.
####2. The amount of space used by an end-post assembly, in feet.
####3. The amount of space between the vines, in feet.

####Once the input data has been entered, the program should calculate and display the number of grapevines that will fit in the row.

```python
print('[[[[[[[[[[[[[[[]]]]]]]]]]]]]]]]')
print('  GRAPEVINE DESIGN CALCULATOR')
print('[[[[[[[[[[[[[[[]]]]]]]]]]]]]]]]\n') 
row_L = float(input(print('Input the length of your rows, in feet: ')))         # Prompt user for the length of row in feet.
end_post_L = float(input(print('Input the amount of space used by an end-post assembly, in feet: '))) # Prompt user for the amount of space for end-post.
spc_bet_vines = float(input(print('Input amount of space between the vines, in feet: '))) # Prompt user for the spacing between vines
grapevines_per_row = ((row_L - (2 * end_post_L)) / spc_bet_vines)               # Calculate number of graevines per row using V=(R-(2*E))/S formula.
print('\nYou can accomidate ', int(grapevines_per_row) , ' grapevines in each row using the following formula: (', row_L, '-(2*', end_post_L, '))/', spc_bet_vines, '=', grapevines_per_row, sep='')  # Print the number of vines per row.
```
