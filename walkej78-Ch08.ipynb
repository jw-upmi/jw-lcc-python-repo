```python
# There are two simple ways to do this.  
# The first replaces Yes with y and is slightly longer:
choice = 'Yes'
choice = choice[0].lower()
if choice == 'y':
    print(choice)
```

```python
# The second is shorter but does not replace Yes with y:
choice = 'Yes'
if choice[0].lower() == 'y':
    print(choice)
```

Write a loop that counts the number of space characters that appear in the string referenced by mystring.

```python
# define string to test
string = 'The Rain in Spain. Add another space.'
# set count to 0
space_count = 0
# define for loop
for l in string:
    # define test
    if l.isspace():
        # if true, add 1 to count
        space_count += 1
# print count        
print(space_count)
```

# PE01. Initials
Write a program that gets a string containing a person’s first, middle, and last names, and displays their first, middle, and last initials. For example, if the user enters John William Smith, the program should display J. W. S.

```python
# Write a program that gets a string containing a person’s first, middle, and last names, and displays their first, middle, and last initials.

def title():
    # print title of program
    print('#####################')
    print('# INITIAL GENERATOR #')
    print('#####################\n')

def initials():
    # define conditional continue
    cont = 'y'
    # create while loop for cont is y
    while cont[0].lower() == 'y':    
        # define initials
        initials = ''
        # prompt user for full name
        fullname_str = input('Please enter a full name: ')
        # strip leading and ending spaces
        fullname_str = fullname_str.strip()
        # conver fullname to name list
        name_list = fullname_str.split()
        # loop to print the first initial in upper case of each name in the list
        for i in name_list:
            initials += i[0] + '. '
        # print results
        print(f'The initials for {fullname_str} are {initials}\n')
        # prompt user to continue
        cont = input('Would you like to find the initials for another name? [y/n]')

# define main
def main():
    # call title
    title()
    # call initials
    initials()

# call main
main()
```
